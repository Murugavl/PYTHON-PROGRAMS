```python
# import the required libraries
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("/content/House_Pice_Dataset.xlsx")

#We using the method <code>head</code> to display the first 5 columns of the dataframe.
df.head()
```

```python
#Fit a linear regression model to predict the 'price' using the feature 'sqft_living' then calculate the R^2.
X = df[['sqft_living']]
Y = df['price']
lm=LinearRegression()
lm.fit(X,Y)
print("Predicting the Price of House for 1500 Square Feet Living:\n")
print(lm.predict([[1500]]))
```

```python
#Fit a linear regression model to predict the 'price' using the list of features.
features =["floors" ,"bedrooms" ,"bathrooms","grade","sqft_living"]
lm.fit(df[features], Y)
print("Predicting the Price of House for the following criteria \n Floors: 2\n Bedrooms: 3\n Bathrooms: 3\n Grade: 7\n Sqft_Living: 2570")
print("\n Price of the House is: ",lm.predict([[2,3,3,7,2570]]))
print("\nSucces rate of the Predicting")
lm.score(df[features], Y)
```
